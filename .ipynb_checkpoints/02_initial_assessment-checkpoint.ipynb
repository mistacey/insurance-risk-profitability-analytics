# Initial Data Assessment

This notebook provides an overview of the raw datasets generated for the project:

- `customers`
- `policies`
- `claims`
- `payments`

We are assessing:

1. Column types and memory usage  
2. Summary statistics for numeric and categorical columns  
3. Missing values 
4. Duplicate rows  

**Note:** Some intentional data issues were introduced during dataset generation:

- Mixed date formats (`YYYY-MM-DD` and `DD/MM/YYYY`)  
- Policies starting before customer join date  
- Outliers in claims and payments  
- Duplicate rows for demonstration  

Logical quality checks (e.g., start date after end date) will be performed in the cleaning notebook after standardizing date formats.

## Imports & Setup

```python
import pandas as pd
import numpy as np

customers = pd.read_csv("data/raw/customers.csv")
policies = pd.read_csv("data/raw/policies.csv")
claims = pd.read_csv("data/raw/claims.csv")
payments = pd.read_csv("data/raw/payments.csv")
```

## Customers Table Assessment

```python
# Basic information
print(customers.info())

# Summary statistics
print(customers.describe(include='all'))

# Missing values
print(customers.isna().sum())

# Duplicate rows
print(customers.duplicated().sum())

# Preview of table
print(customers.head(10))
```

### Observations — Customers

- `join_date` contains mixed formats; some rows may fail parsing.  
- `income` has missing values (to simulate incomplete records).  
- Duplicate rows exist (intentional for demonstration).  

## Policies Table Assessment

```python
# Basic information
print(policies.info())

# Summary statistics
print(policies.describe(include='all'))

# Missing values
print(policies.isna().sum())

# Duplicate rows
print(policies.duplicated().sum())

# Preview of table
print(policies.head(10))
```

### Observations — Policies

- `start_date` and `end_date` are object type  
- No missing values, no duplicates (besides intended logical issues).  
- Logical issues (e.g., policies starting before customer join date, start_date > end_date) will be checked in the cleaning notebook.  

## Claims Table Assessment

```python
# Basic information
print(claims.info())

# Summary statistics
print(claims.describe(include='all'))

# Missing values
print(claims.isna().sum())

# Duplicate rows
print(claims.duplicated().sum())

# Preview of table
print(claims.head(10))
```

### Observations — Claims

- `claim_date` is an object. This must be corrected in the cleanup  
- `claim_type` contains missing values (intentional).  
- Outliers exist in `claim_amount` (simulated extremely high and negative values). 

## Payments Table Assessment

```python
# Basic information
print(payments.info())

# Summary statistics
print(payments.describe(include='all'))

# Missing values
print(payments.isna().sum())

# Duplicate rows
print(payments.duplicated().sum())

# Preview of table
print(payments.head(10))
```

### Observations — Payments

- `payment_date` is object type.  
- No missing values or duplicates.
- No intentional irregularities were introduced to this table.
